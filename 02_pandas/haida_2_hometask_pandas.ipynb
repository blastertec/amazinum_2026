- Finally, load the "Sciamgo Journal and Country Rank data for [Energy Engineering and Power Technology"](http://www.scimagojr.com/countryrank.php?category=2102). It ranks countries based on their journal contributions in the aforementioned area. Call this DataFrame "ScimEn"


- Join the three datasets: Energy, GDP, and ScimEn into a new dataset (using the intersection of country names). Use only the 10 years (2006-2015) of GDP data and only the top 15 countries by Scimagojr 'Rank' (Rank 1 through 15).


- The index of this DataFrame should be the name of the country, and the columns should be<br>

````python
import pandas as pd
import numpy as np

def answer_one() -> pd.DataFrame:
    """
    Loads and cleans Energy, GDP, and ScimEn datasets, then joins them into a
    single DataFrame containing the top 15 countries by Scimagojr Rank
    and GDP data from 2006 to 2015.

    Returns:
        pd.DataFrame: A 20-column DataFrame with 15 entries, indexed by Country.
    """
    energy = pd.read_excel('Energy Indicators.xls', skiprows=17, skipfooter=38)
    energy = energy.iloc[:, 2:]
    energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']

    energy[['Energy Supply', 'Energy Supply per Capita']] = energy[['Energy Supply', 'Energy Supply per Capita']].replace('...', np.nan)
    energy['Energy Supply'] = energy['Energy Supply'] * 1000000

    energy['Country'] = energy['Country'].str.replace(r'\d+', '', regex=True)
    energy['Country'] = energy['Country'].str.replace(r'\s*\(.*?\)', '', regex=True)

    re_names = {
        "Republic of Korea": "South Korea",
        "United States of America": "United States",
        "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
        "China, Hong Kong Special Administrative Region": "Hong Kong"
    }
    energy['Country'] = energy['Country'].replace(re_names)

    GDP = pd.read_excel('world_bank.xls', skiprows=3)

    gdp_re_names = {
        "Korea, Rep.": "South Korea",
        "Iran, Islamic Rep.": "Iran",
        "Hong Kong SAR, China": "Hong Kong"
    }

    GDP['Country Name'] = GDP['Country Name'].replace(gdp_re_names)

    ScimEn = pd.read_excel('scimagojr.xlsx')

    years = [float(y) for y in range(2006, 2016)]
    if not all(y in GDP.columns for y in years):
        years = [int(y) for y in range(2006, 2016)]
    if not all(y in GDP.columns for y in years):
        years = [str(int(y)) for y in range(2006, 2016)]

    GDP_subset = GDP[['Country Name'] + years]
    GDP_subset.columns = ['Country Name'] + [str(int(float(y))) for y in years]

    df = pd.merge(ScimEn[ScimEn['Rank'] <= 15], energy, how='inner', on='Country')
    df = pd.merge(df, GDP_subset, how='inner', left_on='Country', right_on='Country Name')

    df = df.set_index('Country')
    df = df.drop('Country Name', axis=1)

    final_cols = ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations',
                  'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita',
                  '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']

    return df[final_cols]

answer_one()
````

### Question 2
What is the average GDP over the last 10 years for each country? (exclude missing values from this calculation.)

*This function should return a Series named `avgGDP` with 15 countries and their average GDP sorted in descending order.*

````python
def answer_two() -> pd.Series:
    """
    Calculates the average GDP for the top 15 countries over the last 10 years (2006-2015).

    Returns:
        pd.Series: Average GDP for each country, sorted in descending order.
    """
    Top15 = answer_one()
    years = [str(year) for year in range(2006, 2016)]
    avgGDP = Top15[years].mean(axis=1).sort_values(ascending=False)

    return avgGDP.rename('avgGDP')

answer_two()
````

### Question 3
By how much had the GDP changed over the 10 year span for the country with the 6th largest average GDP?

*This function should return a single number.*

````python
def answer_three() -> float:
    """
    Calculates the GDP change over the 10-year span for the country
    with the 6th largest average GDP.

    Returns:
        float: The difference between 2015 GDP and 2006 GDP.
    """
    Top15 = answer_one()
    avgGDP = answer_two()

    target_country = avgGDP.index[5]

    gdp_change = Top15.loc[target_country, '2015'] - Top15.loc[target_country, '2006']

    return gdp_change

answer_three()
````

### Question 4

Create a new column that is the ratio of Self-Citations to Total Citations.
What is the maximum value for this new column, and what country has the highest ratio?

*This function should return a tuple with the name of the country and the ratio.*

````python
def answer_four() -> tuple:
    """
    Calculates the ratio of Self-Citations to Total Citations and finds
    the country with the maximum ratio.

    Returns:
        tuple: (Country name, maximum ratio value)
    """
    Top15 = answer_one()
    Top15['Citation Ratio'] = Top15['Self-citations'] / Top15['Citations']

    max_ratio = Top15['Citation Ratio'].max()
    country = Top15['Citation Ratio'].idxmax()

    return (country, max_ratio)

answer_four()
````

### Question 5

Create a column that estimates the population using Energy Supply and Energy Supply per capita.
What is the third most populous country according to this estimate?

*This function should return a single string value.*

````python
def answer_five() -> str:
    """
    Estimates the population using Energy Supply and Energy Supply per capita,
    and finds the third most populous country.

    Returns:
        str: The name of the third most populous country.
    """
    Top15 = answer_one()
    Top15['Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']

    # Sort population in descending order and get the index of the 3rd one
    third_populous = Top15['Population'].sort_values(ascending=False).index[2]

    return third_populous

answer_five()
````

### Question 6
Create a column that estimates the number of citable documents per person.
What is the correlation between the number of citable documents per capita and the energy supply per capita? Use the `.corr()` method, (Pearson's correlation).

*This function should return a single number.*


````python
def answer_six() -> float:
    """
    Calculates the correlation between citable documents per capita
    and energy supply per capita.

    Returns:
        float: The Pearson correlation coefficient.
    """
    Top15 = answer_one()
    Top15['PopEst'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Citable docs per Capita'] = Top15['Citable documents'] / Top15['PopEst']

    # Calculate Pearson correlation
    correlation = Top15['Citable docs per Capita'].corr(Top15['Energy Supply per Capita'])

    return correlation

answer_six()
````

*This function should return a DataFrame with index named Continent `['Asia', 'Australia', 'Europe', 'North America', 'South America']` and columns `['size', 'sum', 'mean', 'std']`*

````python
def answer_seven(continent_dict: dict) -> pd.DataFrame:
    """
    Groups countries by continent and calculates summary statistics
    for their estimated populations.

    Returns:
        pd.DataFrame: A DataFrame with continent as index and columns
                     ['size', 'sum', 'mean', 'std'].
    """
    Top15 = answer_one()

    Top15['PopEst'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Continent'] = pd.Series(continent_dict)

    return Top15.groupby('Continent')['PopEst'].agg(['size', 'sum', 'mean', 'std'])

ContinentDict  = {'China':'Asia',
                  'United States':'North America',
                  'Japan':'Asia',
                  'United Kingdom':'Europe',
                  'Russian Federation':'Europe',
                  'Canada':'North America',
                  'Germany':'Europe',
                  'India':'Asia',
                  'France':'Europe',
                  'South Korea':'Asia',
                  'Italy':'Europe',
                  'Spain':'Europe',
                  'Iran':'Asia',
                  'Australia':'Australia',
                  'Brazil':'South America'}

answer_seven(ContinentDict)
````
